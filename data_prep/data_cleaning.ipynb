```python
import pandas as pd
import numpy as np
```

```python
df_torii = pd.read_csv('data/IoT-23_Torii.csv')
df_cc = pd.read_csv('data/IoT-23_C&C.csv')
df_normal = pd.read_csv('data/IoT-23_Normal.csv')
```

```python
df_torii["label"] = 1
df_cc["label"] = 1
df_normal["label"] = 0
```

```python
df_torii.head()
```

```python
df_cc.head()
```

```python
df_normal.head()
```

```python
df_attack = pd.concat([df_torii, df_cc], ignore_index=True)
print(df_attack.shape)
print(df_normal.shape)
```

Ensure columns match

```python
print(set(df_attack.columns) - set(df_normal.columns))
print(set(df_normal.columns) - set(df_attack.columns))
```

```python
df_normal_sampled = df_normal.sample(150_000, random_state=42)
print(df_normal_sampled.shape)
```

```python
df_all = pd.concat([df_attack, df_normal_sampled], ignore_index=True)
```

```python
df_all = df_all.drop(['Label'], axis=True)
```

```python
df_all.columns
```

```python
df_selected = df_all[['Flow_Duration', 'Tot_Fwd_Pkts', 'Tot_Bwd_Pkts', 'Pkt_Len_Mean', 'Flow_Byts/s', 'ACK_Flag_Cnt', 'Init_Fwd_Win_Byts', 'SYN_Flag_Cnt', 'label']]
df_selected
```

```python
df_selected["syn_ratio"] = df_selected["SYN_Flag_Cnt"] / (df_selected["Tot_Fwd_Pkts"] + df_selected["Tot_Bwd_Pkts"])
df_selected
```

```python
df_selected.to_csv('../Selected_features.csv')
```
