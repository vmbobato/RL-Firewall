```python
import pandas as pd
```

```python
df = pd.read_csv("../data/preprocessed/Binerized_features.csv")
df.head()
```

```python
df["label"].value_counts(normalize=True)
```

```python
feat = "syn_ratio_bin"

crosstab = (
    df.groupby("label")[feat]
      .value_counts(normalize=True)
      .unstack()
      .fillna(0)
)

print(crosstab)
```

```python
bin_feats = [
    "Flow_Duration_bin",
    "Tot_Fwd_Pkts_bin",
    "Tot_Bwd_Pkts_bin",
    "Pkt_Len_Mean_bin",
    "Flow_Bytes_s_bin",
    "ACK_Flag_Cnt_bin",
    "Init_Fwd_Win_Byts_bin",
    "SYN_Flag_Cnt_bin",
    "syn_ratio_bin",
]

for feat in bin_feats:
    print(f"\n=== {feat} ===")
    crosstab = (
        df.groupby("label")[feat]
          .value_counts(normalize=True)
          .unstack()
          .fillna(0)
    )
    print(crosstab)
```
