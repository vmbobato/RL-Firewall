## Putting the Data into Bins

```python
import pandas as pd
import numpy as np
```

```python
df_rl = pd.read_csv('../data/preprocessed/Selected_features.csv')
```

```python
len(df_rl)
```

```python
df_rl.head()
```

```python
df_rl.columns
```

```python
df_rl[[
    "Flow_Duration", "Tot_Fwd_Pkts", "Tot_Bwd_Pkts",
    "Pkt_Len_Mean", "Flow_Byts/s", "ACK_Flag_Cnt",
    "Init_Fwd_Win_Byts", "SYN_Flag_Cnt", "syn_ratio"
]].describe()
```

```python
df_rl["Flow_Duration_bin"] = pd.qcut(
    df_rl["Flow_Duration"],
    q=5,
    labels=False,
    duplicates="drop"
)
```

```python
df_rl["Tot_Fwd_Pkts_bin"] = pd.cut(
    df_rl["Tot_Fwd_Pkts"],
    bins=[-1, 1, 3, 10, 100, 1000, float("inf")],
    labels=False
)
```

```python
df_rl["Tot_Bwd_Pkts_bin"] = pd.cut(
    df_rl["Tot_Bwd_Pkts"],
    bins=[-1, 0, 1, 5, 50, float("inf")],
    labels=False
)
```

```python
df_rl["Pkt_Len_Mean_bin"] = pd.cut(
    df_rl["Pkt_Len_Mean"],
    bins=[-0.1, 0.1, 200, 800, float("inf")],
    labels=False
)
```

```python
df_rl["Flow_Bytes_s_bin"] = pd.cut(
    df_rl["Flow_Byts/s"],
    bins=[-1, 1, 1000, 10000, 1000000, float("inf")],
    labels=False
)
```

```python
df_rl["ACK_Flag_Cnt_bin"] = pd.cut(
    df_rl["ACK_Flag_Cnt"],
    bins=[-1, 0, 1, 3, 10, float("inf")],
    labels=False
)
```

```python
df_rl["Init_Fwd_Win_Byts_bin"] = pd.qcut(
    df_rl["Init_Fwd_Win_Byts"],
    q=5,
    labels=False,
    duplicates="drop"
)
```

```python
df_rl["SYN_Flag_Cnt_bin"] = pd.cut(
    df_rl["SYN_Flag_Cnt"],
    bins=[-1, 0, 1, 3, 5, 10, float("inf")],
    labels=False
)
```

```python
df_rl["syn_ratio_bin"] = pd.cut(
    df_rl["syn_ratio"],
    bins=[-0.01, 0.01, 0.5, 0.9, 1.0, 1.01],
    labels=False
)
```

```python
df_env = df_rl[
    [
        "Src_IP",
        "Flow_Byts/s",
        "Flow_Duration_bin",
        "Tot_Fwd_Pkts_bin",
        "Tot_Bwd_Pkts_bin",
        "Pkt_Len_Mean_bin",
        "Flow_Bytes_s_bin",
        "ACK_Flag_Cnt_bin",
        "Init_Fwd_Win_Byts_bin",
        "SYN_Flag_Cnt_bin",
        "syn_ratio_bin",
        "label"
    ]
].reset_index(drop=True)
```

```python
bin_cols = [
    "Flow_Duration_bin",
    "Tot_Fwd_Pkts_bin",
    "Tot_Bwd_Pkts_bin",
    "Pkt_Len_Mean_bin",
    "Flow_Bytes_s_bin",
    "ACK_Flag_Cnt_bin",
    "Init_Fwd_Win_Byts_bin",
    "SYN_Flag_Cnt_bin",
    "syn_ratio_bin",
]

df_env = df_env.dropna(subset=bin_cols + ["label"]).reset_index(drop=True)
```
